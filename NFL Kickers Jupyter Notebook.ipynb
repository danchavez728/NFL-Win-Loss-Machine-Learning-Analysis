```python
#Dependencies
from matplotlib import pyplot as plt
import sklearn.datasets as dta
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Load Kicker data set into pandas
Kicker_data = pd.read_csv("Resources\Kicker_Fantasy.csv")
Kicker_data.head()
```

```python
# Create a group based on the values in the 'Player' column
Player_group = Kicker_data.groupby('FG%')

# Count how many times each maker appears in our group
Player_makers = Player_group['FG%'].count()

Player_makers
```

```python
# Create a bar chart based off of the group series from before
count_chart = Player_makers.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Field Goal Percentage")
count_chart.set_ylabel("Number of Players")


plt.show()
plt.tight_layout()
```

```python
# Set the 'Player' as our index
Week_index = Kicker_data.set_index('Week')
Week_index.head()
```

```python
# Create a scatter plot which compares MPG to horsepower
Week_index.plot(kind="scatter", x="Player", y="FG%", grid=True, figsize=(8,8),
              title="Player vs. Field Goal Percentage")
plt.show()
```

```python
# Generate a scatter plot of year versus violent crime rate
Player = Kicker_data.iloc[:,0]
Field_Goal_Percentage = Kicker_data.iloc[:,6]
plt.scatter(Player,Field_Goal_Percentage)
plt.xticks(Player, rotation=90)
plt.xlabel('Player')
plt.ylabel('Field Goal Percentage')
plt.show()
```
