```python
# Load Kicker data set into pandas
Kicker_data = pd.read_csv("Resources\Kicker_Fantasy.csv")
Kicker_data.head()
```

```python
# Create a group based on the values in the 'Player' column
Player_group = Kicker_data.groupby('FG%')

# Count how many times each maker appears in our group
Player_makers = Player_group['FG%'].count()

Player_makers
```

```python
# Create a bar chart based off of the group series from before
count_chart = Player_makers.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Field Goal Percentage")
count_chart.set_ylabel("Number of Players")

#Save the plotted figures as .pngs
#plt.savefig("Kicker_Bar.png")

#plt.show()
plt.tight_layout()
#plt.show()

#Save the plotted figures as .pngs
plt.savefig("Kicker_Bar.png")
```
