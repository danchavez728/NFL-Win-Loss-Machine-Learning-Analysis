```python
#Dependencies
from matplotlib import pyplot as plt
import sklearn.datasets as dta
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Load Kicker data set into pandas
Kicker_data = pd.read_csv("Resources\Kicker_Fantasy.csv")
Kicker_data.head()
```

```python
#Kicker_data["FG%"].astype(float).mean()
#Kicker_data['FG%'] = Kicker_data['FG%'].str.rstrip('%').astype('float') / 100.0
```

```python
# Create a group based on the values in the 'Player' column
Player_group = Kicker_data.groupby('FG%')

# Count how many times each maker appears in our group
Player_makers = Player_group['FG%'].count()

Player_makers
```

```python
# Create a bar chart based off of the group series from before
count_chart = Player_makers.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Field Goal Percentage")
count_chart.set_ylabel("Number of Players")

#Save the plotted figures as .pngs
#plt.savefig("Kicker_Bar.png")

#plt.show()
plt.tight_layout()
#plt.show()

#Save the plotted figures as .pngs
plt.savefig("Kicker_Bar.png")
```

```python
# Load Kicker data 1 set into pandas
Kicker_data2 = pd.read_csv("Resources\Kicker_Fantasy (1).csv")
Kicker_data2.head()
Kicker_data.dropna()
```

```python
#set Play.1 Index

Kicker_data3 = Kicker_data2.set_index("Player.1")
Kicker_data3.head(10)
```

```python
# extracting greatest 10 
large10_df = Kicker_data3["FF Season Ave"].nlargest(10) 
  
# display 
large10_df.head(10)
```

```python
# Create a bar chart based off of the group series from before
top_10_chart = large10_df.plot(kind='bar')

# Set the xlabel and ylabel using class methods
top_10_chart.set_xlabel("Player.1")
top_10_chart.set_ylabel("FF Season Ave")

plt.tight_layout()

plt.savefig("Kicker_Top_10.png")

#plt.tight_layout()

#plt.savefig("Kicker_Top_10.png")
```

```python
#Create a Line Chart (?)
large10_df.plot(label="Fantasy Football Season Average")
large10_df.loc['Player.1'].plot(label="Top 10 Players")
plt.legend()

plt.savefig("Kicker_Top10_line.png")

plt.show()


#plt.savefig("Kicker_Top10_line.png")
```
