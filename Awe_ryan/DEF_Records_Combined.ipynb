```python
import pandas as pd
from glob import glob
import numpy as np

```

```python
recorddf = pd.read_csv ('HA.csv')
```

```python
def_data_combined = sorted(glob('fantasy-stats-dst-*.csv'))
```

```python
df = pd.concat((pd.read_csv(file).assign(Year = file)
        for file in def_data_combined), ignore_index = True)
```

```python
df.replace(to_replace ="fantasy-stats-dst-2017.csv", value ="2017")
df.replace(to_replace ="fantasy-stats-dst-2018.csv", value ="2018")
df.replace(to_replace ="fantasy-stats-dst-2019.csv", value ="2019")
```

```python
win_loss = recorddf[['score_home','score_away','team_home','team_away','schedule_date','schedule_season','schedule_week']]
```

```python
def win_df(win_loss):

    if (win_loss['score_home'] <= win_loss['score_away']):
        return 'loss'
    elif (win_loss['score_home'] > win_loss['score_away']):
        return 'win'
    

win_loss['wins_losses'] = win_loss.apply(win_df, axis = 1)

win_loss['total_score'] = win_loss['score_home'] + win_loss['score_away']

win_loss
```
